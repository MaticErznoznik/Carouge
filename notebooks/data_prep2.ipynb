```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras import backend as K
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
# load partial feature vectors and append FVs of correct length (Len timesteps)

x_train = []
y_train = []
lens = []
Len = 80
for i in ['flowerbed2.npy', 'flowerbed3.npy', 'flowerbed7.npy', 'flowerbed8.npy']:
    data = np.load(i,allow_pickle = True)

    for j in range(len(data)):
        if(len(data[j][0]) > Len):    
            for k in range(len(data[j][0])-Len):
                if(k<5):
                    y_train.append(np.array(data[j][0][1 + k:Len + k]))
                    x_train.append([data[j][0][1 + k], data[j][2], datetime.fromtimestamp(data[j][1][0]).hour])
                    if(np.isnan(data[j][0][0])):
                        y_train = y_train[:-1]
                        x_train = x_train[:-1]
                    lens.append(len(y_train[-1]))
        

x_train = np.array(x_train)
y_train = np.array(y_train)

d = np.array([0, 0], dtype = object)
d[0] = x_train
d[1] = y_train


#normalize (the model's output should be multiplied by 30 to get accurate values)
x_train /= 30
y_train /= 30          


shuffler = np.random.permutation(len(x_train))
x_train = x_train[shuffler]
y_train = y_train[shuffler]
print(x_train)


d = np.array([0, 0], dtype = object)
d[0] = x_train[:,:7]
d[1] = y_train

np.save('train_data_carouge_updated.npy',d, allow_pickle = True)
```

```python
#load complete feature vectors

x_train, y_train = np.load(f'train_data_carouge_updated.npy', allow_pickle = True)

shuffler = np.random.permutation(len(x_train))
x_train = x_train[shuffler]
y_train = y_train[shuffler]

y_train = y_train
x_train = x_train


print(len(y_train[0]))

#show moistures distribution
plt.hist(y_train[:,0]*30, bins = 50)
plt.show()
```

```python
#Build and train model

Len = 79
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(7,activation='relu'))
model.add(tf.keras.layers.Dense(100))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(200, activation='tanh'))
model.add(tf.keras.layers.Dropout(0.2))
#model.add(tf.keras.layers.Dense(50, activation='tanh'))

model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(Len, activation='relu'))

model.compile(optimizer =tf.keras.optimizers.Adam(lr = 0.001, beta_1 = 0.99), loss = 'mse')

batch_size = 50
history = model.fit(x_train[10:][:, :7],y_train[10:], epochs =300, batch_size = batch_size, validation_data = (x_train[:10][:, :7], y_train[:10]), verbose = 1)

plt.plot(history.history['loss'])
plt.yscale('log')
```

```python
#visualise results compared to truth

for test in [7, 8, 9]:
    y_pred = model.predict(np.atleast_2d(x_train[test][:7]))

    #adjust, so the prediction for t0 is equal to current moisture
    y_pred[0] = np.add(y_pred[0], (+y_train[test][0] - y_pred[0][0]))
    print(30*y_pred[0][-1] - 30*y_pred[0][0])

    plt.plot(30*y_pred[0], label = 'prediction')
    plt.plot(30*y_train[test], label = 'true')



plt.xlabel('intervals')
plt.ylabel('moisture[%]')

```
